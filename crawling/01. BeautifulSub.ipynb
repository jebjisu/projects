### 뉴스데이터 수집해보기
- 여러 페이지의 내용을 수집해야 할 경우
- 눈에 보이지 않는 데이터 수집 필요 -> a(href:주소)

데이터 수집 순서
1) 뉴스 페이지의 url주소로 문서를 요청
2) 응답받은 웹 문서를 변환(파싱)
3) 변환된 웹 문서에서 url을 수집
4) 수집된 url에 접근해서 제복, 내용 수집
5) 저장하기

```python
#1) url, request
url = 'https://news.naver.com/section/105'
res = req.get(url) #위에 유알엘 가지고 와주세요 요청 
res.text

#2) Lxml
html = bs(res.text, 'lxml')
```

```python
#1) 뉴스페이지의 url주소 수집 및 리스트로 저장하기
news_list = html.select('a.sa_text_title')
news_list
```

```python
news_list[0].text
```

```python
#반복문 활용해서 내용 가져오기 

for i in range(len(news_list)):
    r = news_list[i].text
    print(r)
```

```python
#url리스트에 저장된a요소에서 href 속성 값을 리스트로 저장하기
news_list[0].get('href')
```

```python
#반복문 for문을 활용해서, hrdf 리스트 모아보기 

hred_list = []
for i in range(len(news_list)):
    hred_list.append(news_list[i].get('href'))
    
hred_list
```

```python
#수집된 url에 접근해서 제목, 내용 수집하기 

for herf in hred_list:
    #url을 요청하겠습니다. 
    uews_url1 = herf
    res = req.get(uews_url1)

    #lxml파싱
    html = bs(res.text, 'lxml')

    #title/contnet
    news_list = html.select('a.sa_text_title')

    #print
    print('뉴스제목:', title)
    print('내용', content)
    print('='*50)
```

```python
##강사님 코드 
 #4) 수집된url에 접근해서 제목, 내용 수집하기
for href in hred_list:
    res=req.get(href)
    html=bs(res.text, 'lxml')

    title=html.select_ont(' #title_are>span').text
    content = html.select_one('#dic_Area').text.strip()

    #print
    print('뉴스제목:', title)
    print('내용', content)
    print('='*50)
    
```

```python
data_list = []
# 4) 수집된 url에 접근해서 제목,내용 수집하기

for href in hred_list:
    # url 요청
    res = req.get(href)
    
    # lxml파싱
    html = bs(res.text , 'lxml')
    
    # title/content
    title = html.select_one('#title_area>span').text
    content = html.select_one('#dic_area').text.strip()
    
    data_list.append([title, content,href])
```

```python
import pandas as pd 
```

```python
#강사님 코드
url = 'https://search.zdnet.co.kr/?kwd=%EC%9D%8C%EC%8B%9D%EB%AC%BC'

res=req.get(url)

html = bs(res.text, 'lxml')

aTags = html.select('.news_box .assetText a')
aTags

```

```python
titles = html.select('.assetText h3')
titles
```

```python
news_post = []
for i in range (len(titles)):
    news_post.append([titles[i].text, aTags[i].get("href")])

news_post
```

```python
data_list = []

for title, href in news_post:
    res = req.get(href)
    html = bs(res.text,'lxml')
    content = html.select_one('#articleBody').text.strip()
    data_list.append([title,content,href])

news_df= pd.DataFrame(data_list, columns =['뉴스제목', '내용', '페이지주소'])
news_df
```
