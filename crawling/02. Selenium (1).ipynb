```python
#1. 웹 브라우저를 자동으로 실행하기 위한 기능을 제공함 - 크롬,엣지, 사파리 등 다양한 브라우저를 지원
from selenium import webdriver as wb

#2. HTML 요소를 찾기 위한 도구를 제공함
#요소 찾는 법 : ID, 클래스명, 태그명, CSS선택자, XPATH등 
from selenium.webdriver.common.by import By

#3. 컴퓨터 키보드의 값을 제공하는 역할을 한다. - 사람이 직접 키보드를 눌러야 하는 상황을 코드로 대체할 때 사용한다. 
from selenium.webdriver.common.keys import Keys
```

```python
#1.브라우저 실행/페이지 요청하기 
driver = wb.Chrome()
#재사용성을 위해 driver 변수에 넣어줌
#Chrome() : 
```

```python
#페이지 요청하기
driver.get('https://www.naver.com/')
```

```python
# 검색창 요소에 접근하기 : ID 값을 사용해보기 
query_input = driver.find_element(By.ID, "query")

# 접근한 요소에 키워드를 전송하기
query_input.send_keys("음식물")
#send_keys(): 찾은 요소에 키보드 입력을 자동으로 전달하는 메서드

#실제 키보드에서 엔터치기
query_input.send_keys(Keys.ENTER)
#Keys.Enter: 실제 키보드에서 enter를 누르는 것 처럼 작동을 함
```

```python
#스크롤 해보기
body = driver.find_element(By.TAG_NAME,"body")
body.send_keys(Keys.END)
```

```python
# 드라이버 뒤로가기
driver.back()
#back()은 WebDriver객체의 메서드로 현재 열려 있는 브라우저에서 뒤로가기 버튼을 누른 것과 동일함. 
```

```python
#드라이버 종료하기
driver.close()
#close(): 열려있는 브라우저를 종료할 때 사용하는 메서드 
#quit() : 열려있는 브라우저 창을 종료하는 것 뿐만 아니라 Selenium WebDriver 세션까지도 종료 (완전히 완료)
```

## 네이버 뉴스데이터 수집 실습 
1. 네이버 검색 페이지 실행하기
2. 스크롤 한번 해서, '기사 더보기 ' 버튼 클릭하기
3. 뉴스  URL 요소 수집하기
4. 수집된 URL에서 제목, 내용을 반복문을 활용해서 수집하기
5. DataFrame 으로 만들어서 엑셀 파일로 저장하기

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
url = 'https://news.naver.com/section/105'
```

```python
#Chrome() 활용해서 모달창 띄우기 (webdriver)
driver = wb.Chrome()
```

```python
#1. url로 이동하기 (get)
```

```python
driver.get('https://news.naver.com/section/105')
```

```python
#Chrome 창 최대화하기 
driver.maximize_window()
#maximize_window() : WebDriver 가 띄운 브라우저 창을 사용자 화면에 맞게 최대 크기로 확장해준다. 
```

```python
#스크롤하기 
body = driver.find_element(By.TAG_NAME,"body")
body.send_keys(Keys.END)
```

```python
#기사 더보기 클릭하기 - XPATH 사용해보기
#XPATH : HTML 문서에 있는 특정 요소를 찾기 위한 경로임
#복잡한 조건이나, 정확한 위치 지정이 필요할 때 사용합니다

button = driver.find_element(By.XPATH, '//*[@id="newsct"]/div[4]/div/div[2]/a')
button.click()
#click() : Selenium에서 웹요소 (벝ㄴ, 링크 등)을 마우스로 클릭하는 동작을자동으로 수행하는 메서드 
```

```python
#3. 뉴스 URL 수집하기
# 3-1. a태그 : CLASS_NAME으로 요소 선택해보기
titles = driver.find_elements(By.CLASS_NAME , 'sa_text_title._NLOG_IMPRESSION')
titles

#Selenium 은 글자를 뽑아주는 것이 아니고, 이건 선택 내용이 있는 HTML요소다. 라고 알려줌
#.text
```

```python
titles[0].text
```

```python
#for문을 활용하여 titles에 있는 제목 모두 출력해보기
#1) range 사용하여 추출하는 방법
for i in range(len(titles)):
    print(titles[i].text)
```

```python
#첫번째 a태그의 요소인 href 속성 값 가져오기
titles[0].get_attribute("href")

#get_attribute() : HTML 태그 안에 있는 특정 속성값을 가져오는 메서드 이다. 
```

```python
#for문을 활용해서 titles 에 있는 모든 a태그의 hredf 속성 값 저장해보기
href_list= []

for i in range(len(titles)):
    href_list.append(titles[i].get_attribute("href"))
href_list
```

```python
#4. 수집한 URL에서 제목 내용 수집하기
#4-1) 첫번째 기사 클릭해보기
titles[1].click()
```

```python
#4-2) 제목 요소 선택해서 텍스트로 출력해보기 - ID로 선택해보기!
title = driver.find_element(By.ID,"title_area")
title.text
```

```python
#4-3) 내용 요소를 선택하여 텍스트로 출력해보기 - Class
content = driver.find_element(By.CLASS_NAME, 'newsct_article._article_body')
content.text
```

```python
#href_list에 접근해서 title,content를 모아주기 
# 1) url 주소로 웹 문서를 요청하기 (새로운 페이지를 로딩해준다)

# 2) 제목, 내용 요소 선택해서, 컨텐츠를 추출하기


# 3) 빈 리스트에 append 해주기 


df = []

for href in href_list:
    # 1) URL 주소로 웹문서를 요청하기 (새로운 페이지를 로딩)
    driver.get(href)

    # 2) 제목과 내용 요소를 선택해서, 컨텐츠를 추출하기
    title = driver.find_element(By.ID, 'title_area').text
    content = driver.find_element(By.CLASS_NAME, 'newsct_article._article_body').text

    # 3) 빈 리스트에 append 해주기
    df.append([title, content, href])

# 드라이버 닫기 
driver.close()

df
```

```python
#데이터 프레임으로 만들기 
import pandas as pd

new_df = pd.DataFrame(df, columns =['뉴스 제목','내용','URL'])
new_df
```

```python
# 4) 엑셀로 저장하기 
new_df.to_excel("IT과학 뉴스데이터 수집.xlsx")
```
