```python
import pandas as pd
df = pd.read_excel('IT과학 뉴스데이터 수집.xlsx')
df
#2) 내용 전처리하기 

df['내용'] = df['내용'].str.replace('\n','')
df['내용'] = df['내용'].str.replace('\t','')
df
```

### 뉴스 본문에서(내용) 명사만 추출하기 

```python
def extract_nouns(text):
    nouns = [] #명사를 저장할 빈 리스트를 만들기 
    
    #okt-> (말, noun)
    for word, tag in okt.pos(text):
        if tag == "Noun":
            nouns.append(word) #명사리스트 반환
    return nouns 
```

```python
#apply() : series나 DataFrame 의 각 값에 함수를 적용해 결과를 반환하는 메서드
df["nouns"] = df["내용"].apply(extract_nouns)
```

```python
df["nouns"]
```

# 3) 단어별 빈도 계산하기

- counter : 단어의 빈도수를 카운팅해주는 도구

```python
from collections import Counter
```

```python
all_nouns = []

for nouns in df["nouns"]:
    all_nouns.extend(nouns)
    #extend : 리스트 메서드로, 여러 요소를 한번에 기존 리스트에 추가할 떄 사용 
    #append -> 리스트 전체를 하나의 요소로 추가하고, extend는 각 요소를 풀어서 하나씩 리스트에 이어붙임. ''

#단어별로 몇 번 나왔는지 ㅋ카운팅해서, 딕셔너리 형태로 저장
freq_dict =Counter(all_nouns)
```

```python
#가장 많이 등장한 10개 단어를 출력해보기 
#most_common( ) : 빈도수가 가장 높은 순서대로 상위 n개의 단어와 그 횟수를 뽑아주는 함수

top10 = freq_dict.most_common(10)
print ('상위10개 단어:', top10)
```

```python
# 명사를 추출하는 함수에, 단어 길이가 2 이상인 것만 반환하는 코드를 수정 해보기
def extract_nouns(text):
    nouns = [] # 명사를 저장할 빈 리스트

    #okt -> (말,Noun)
    for word, tag in okt.pos(text):
        if tag == "Noun" and len(word) >= 2:
            nouns.append(word) 
    return nouns #명사리스트 반환
    
```

```python
#2) 전처리된 텍스트에서 명사목록 만들어 새 컬럼에 저장하기
df["nouns"] = df["내용"].apply(extract_nouns)
df
```

```python
#3) counter
all_nouns = []

for nouns in df["nouns"]:
    all_nouns.extend(nouns)
    #extend : 리스트 메서드로, 여러 요소를 한번에 기존 리스트에 추가할 떄 사용 
    #append -> 리스트 전체를 하나의 요소로 추가하고, extend는 각 요소를 풀어서 하나씩 리스트에 이어붙임. ''

#단어별로 몇 번 나왔는지 ㅋ카운팅해서, 딕셔너리 형태로 저장
freq_dict =Counter(all_nouns)
```

```python
#4) top10
top10 = freq_dict.most_common(10)
print ('상위10개 단어:', top10)
```

#### 워드 클라우드 그리기


```python
wc = WordCloud(
    font_path ='C:/Windows/Fonts/malgunbd.ttf', #한글이 꺠지지 않도록 폰트 경로 설정
    background_color = 'white',
    width = 600, height = 300,
).generate_from_frequencies(dict(freq_dict))
#generate_from_frequencies( ): 단어- 빈도 딕셔너리를 바탕으로, 워드클라우드 생성 

#2) matplotlib활용하여 워드클라우드 표시하기
plt.figure(figsize=(10,8))
plt.imshow(wc, interpolation = 'bilinear')#이미지를 시각화 하는 #bilinear이미지를 부드럽게 해주는 옵션 
plt.axis('off') #축을 제거하기
```
