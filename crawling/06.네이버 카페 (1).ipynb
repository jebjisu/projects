### URL 수집하기 
1. url 가지고 있는 a태그 찾아보기 -> CSS_SELSECTOR로
2. url_list 빈 리스트 만들기
3. href 속성 가져오는 반복문 실행


```python
cafe_titles = driver.find_elements(By.CSS_SELECTOR,'div > a.article')
#get.attribute('href')
cafe_titles
```

```python
href_list = []

for i in cafe_titles:
    href = i.get_attribute('href')
    href_list.append(href)
href_list
```

```python
len(href_list)
```

```python
#두번째 페이지 url도 가지고오기 
second_page_titles = driver.find_elements(By.CSS_SELECTOR,'a.article')
second_page_titles
```

```python
#page_url_list -> 100개
page_url_list = []

#첫번째 페이지 


for i in second_page_titles:
    href= i.get_attribute('href')
    page_url_list.append(href)

page_url_list  
```

```python
all_page_list =(href_list + page_url_list )
```

```python
#한개씩 하지말고, f-string
f_url = []
driver=wb.Chrome()

for i in range(1,5):
    url = f'https://cafe.naver.com/f-e/cafes/10298136/menus/23?viewType=L&page={i}&size=50&ta=SUBJECT&q=%EC%9D%8C%EC%8B%9D%EB%AC%BC%EC%B2%98%EB%A6%AC%EA%B8%B0'
    driver.get(url)
    #time.sleep(3)
    driver.implicitly_wait(10)
    aTags = driver.find_elements(By.CSS_SELECTOR, 'div > a.article')
    for tag in aTags:
        href = tag.get_attribute('href')
        f_url.append(href)

driver.quit()
```

```python
#합치기 

dates = []
titles =[]
contents =[]

for url in page_url_list[:10]:
    driver.get(url)
    time.sleep(1)

    driver.switch_to.frame('cafe_main')
    title = driver.find_element(By.CSS_SELECTOR,'.title_text').text
    titles.append(title)

    date = driver.find_element(By.CLASS_NAME,'date').text
    dates.append(date[:-7])

    content=driver.find_element(By.CLASS_NAME,'se-module').text
    contents.append(content)

```

```python
import pandas as pd

data = {
    'title':titles,
    'date':dates,
    'contents':contents
}

df = pd.DataFrame(data)
```
