```python
#a 태그를 찾아서, href 속성 추출 -- CSS_SELECTOR
blind_titles = driver.find_elements(By.CSS_SELECTOR, 'div.tit > h3 > a')
```

```python
#a태그를 순회하며, href속성 추출해서 list에 append
href_list = []

for i in blind_titles:
    href = i.get_attribute('href')
    href_list.append(href)

href_list
```

```python
#4) 댓글 1개 가져오기 CSS_SELECTOR
comment = driver.find_element(By.CSS_SELECTOR,'.cmt-txt')
comment.text

#모든 댓글을 담고싶을 땐?
#comment = driver.find_elements(By.CSS_SELECTOR,'wrap-comment.comment_area > p.cmt-txt')
```

```python
#5) 모든 댓글 가져오기 
#5-1) 모든 댓글 요소 가져오기 
comment_blocks = driver.find_elements(By.CSS_SELECTOR, 'div.wrap-comment.comment_area')

comments = []

for block in comment_blocks:
    text = block.find_element(By.CSS_SELECTOR, 'p.cmt-txt').text
    comments.append(text)

comments
                                
```

```python
# href_list에 있는 모든 내용에 대해 제목,, 내용, 댓글, 날짜 가져오기

# 빈 리스트에 append까지 완료하기
 
data_list = []
 
 
for href in href_list[:10]:

    # 1) 페이지 요청하고, time.sleep으로 로딩 대기

    driver.get(href)

    time.sleep(2)

    # 2) 제목 추출하기

    title = driver.find_element(By.TAG_NAME, 'h2').text

    # 3 ) 본문 추출하기

    content = driver.find_element(By.CSS_SELECTOR, '#contentArea').text

    # 4) 작성일 추출하기

    date = driver.find_element(By.CSS_SELECTOR, 'span.date').text

    # 5) 댓글 전체 추출하기

    comment_blocks = driver.find_elements(By.CSS_SELECTOR, 'div.wrap-comment.comment_area')

    comments = []

    for block in comment_blocks:

        text = block.find_element(By.CSS_SELECTOR, 'p.cmt-txt').text

        comments.append(text)
 
    # 6) data_list에 append해주기

    data_list.append([title, content, comments, date, href])


# 7) 드라이버 종료

driver.quit()
 
data_list

 
```

```python
# 데이터 프레임으로 만들기
import pandas as pd

blind_df = pd.DataFrame(data_list, columns = ['title', 'content', 'comments', 'date', 'url'])
blind_df.head()
```

```python
#csv 파일로 저장하기 
blind_df.to_csv('블라인드_음식물처리기_10개만.csv', encoding = 'utf-8-sig')
```
