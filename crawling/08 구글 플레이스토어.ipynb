### JavaScript 활용하여 스크롤하기 

```python
# 1. 모댤 높이 가져오기 

#document.querySelector : 모달 요소 선택 
#scrollHeight : 전체 높이 ,  화면에 보이지 않는 스크롤 영역의 높이포함
#driver.execute_script :  자바스크립트 실행 

old_height = driver.execute_script('''
    const modal = document.querySelector('div.odk6He')
    return modal.scrollHeight

''')

print(f'모달높이:{old_height}')
```

```python
#스크롤 라이르버리 
# click (), send_keys() 같은 단일 명령보다 더 정교한 동작을 사용할 때 사용 (더블 클릭, 마우스 이동과 같은 정교한 역할을 한다)
from selenium.webdriver.common.action_chains import ActionChains 

#스크롤을 어디에서 시작할지 정의하고, 화면을 위아래로 움직일 수 있음 
#마우스 내리게 하는 동작 라이브러리 
from selenium.webdriver.common.actions.wheel_input import ScrollOrigin
```

```python
# 1) 모달 요소 선택 -- CSS_SELECTOR
modal = driver.find_element(By.CSS_SELECTOR, 'div.odk6He')

# 2) 모달요소의 기준점 잡고, ScrollOrigin 생성하기
scroll_origin = ScrollOrigin.from_element(modal)

# 3) ActionChains 객체만들기
actions = ActionChains(driver)

# 4) 스크롤 실행 
# scroll_from_origin(스크롤 시작기준이 되는 요소,가로로 스크롤 픽셀값, 세로로 스크롤할 픽셀값)
actions.scroll_from_origin(scroll_origin,0,1000).perform()

print('스크롤 완료')
```

```python
#3단계 

#1. 초기 높이 가져오기
old_height = driver.execute_script('''
    const modal = document.querySelector('div.odk6He')
    return modal.scrollHeight

''')

# 변수
count = 0
total_scrolls = 5
height = 0

#2. WHILE 반복문 활용
while count < total_scrolls :
    height +=20000

    #지정한 orirgin 기준으로 height만큼 스트롤 
    actions.scroll_from_origin(scroll_origin,0,height).perform()
    time.sleep(1)


    #스크롤 후 모달의 높이 가져오기 execute_script
    new_height = driver.execute_script('''
    const modal = document.querySelector('div.odk6He')
    return modal.scrollHeight

''')

    if new_height == old_height:
        break
        
    old_height = new_height
    count +=1
print('스크롤 완료')
#3. 초기높이 -> 새로운 높이 갱신해서 다음 반복
#4. 5번만 
```
