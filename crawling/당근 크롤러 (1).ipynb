```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
URL = "https://www.daangn.com/kr/community/?in=%EB%8F%99%EA%B5%90%EB%8F%99-6098&search=%EB%9F%AC%EB%8B%9D"
MAX_CLICKS = 3 # 더보기 클릭 최대 횟수
PAUSE = 1.0 # 스크롤/클릭 사이 대기(초)
```

```python
driver = webdriver.Chrome() # 크롬드라이버가 PATH에 있어야 합니다
driver.get(URL)
```

```python
wait = WebDriverWait(driver, 5)

for i in range(MAX_CLICKS):
# 바닥까지 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(PAUSE)
    print(f"[{i+1}] 스크롤 완료")
# 더보기 버튼 클릭 시도
    try:
        more_btn = wait.until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(., '더보기')]"))
    )
    except Exception:
        print("더보기 버튼을 찾지 못해 종료합니다.")
        break

    driver.execute_script("arguments[0].scrollIntoView({block:'center'});", more_btn)
    driver.execute_script("arguments[0].click();", more_btn)
    time.sleep(PAUSE)
    print(f"{i+1}회 클릭 완료")
```

```python
xp = (
"//a"
"[contains(concat(' ', normalize-space(@class), ' '), ' _20wyd00 ')"
" and contains(concat(' ', normalize-space(@class), ' '), ' sprinkles_display_flex_base__1byufe82i ')"
" and contains(concat(' ', normalize-space(@class), ' '), ' sprinkles_gap_1.5_base__1byufe8s6 ')"
" and contains(concat(' ', normalize-space(@class), ' '), ' sprinkles_gap_4_medium__1byufe8qs ')]"
)

els = wait.until(EC.presence_of_all_elements_located((By.XPATH, xp)))
```

```python
hrefs = []
seen = set()
for e in els:
    h = e.get_attribute("href")
    if h and (h not in seen):
        seen.add(h)
        hrefs.append(h)

print(f"수집 개수: {len(hrefs)}")
print(hrefs[:10]) # 샘플 확인
```

```python
title_el = driver.find_element(By.XPATH, '//*[@id="main-content"]/div[1]/div/div/div/div[1]/h1')
title = title_el.text.strip()
print("제목:", title)
```

```python
time_el = driver.find_element(By.XPATH, '//*[@id="main-content"]/div[1]/div/div/div/div[1]/div[1]/button/div/span/span/time')
date_iso = time_el.get_attribute("datetime")
date = date_iso[:10]
print("날짜(datetime):", date)
```
