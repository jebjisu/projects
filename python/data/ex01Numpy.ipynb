### BMI 값 구하기 
- txt 파일 불러오는 방법 알기
- numpy 배열 연산 진행하기
- 1m = 100cm /(cm단위 /100 -> m)
- 1kg = 2.204pound (pound / 2.204 -> kg)

```python
# txt를 불러오는 함수 : Loadtxt()
# 외부 파일을 불러올 때는 "경로/파일명.확장자" 정확하게 기록
data = np.loadtxt('data/height_weight(p).txt', delimiter = ",")
data
#delimite: 구분 기호 
```

```python
#bmi지수 구하기 
height_cm = data[0]
weight_p = data[1]

#데이터를 키와 몸무게 정보로 나누기 

#1cm = 0.01m ->키를 m 단위로 바꾸기
height_m = height_cm / 100
#2.204kg = 1pound -> 몸무게를 kg 단위로 바꾸기
weight_kg = weight_p / 2.204
#BMI지수 구하기 
BMI = weight_kg/ height_m*height_m
BMI
```

```python
name_score = np.array([['지원',100],
                       ['민철',90],
                       ['기준',80],
                       ['지희',70]])

name_score
```

```python
#민철의 정보를 가져오기 -> =='민철'
#1)  이름 정보만 가져오기 
name_score [ : ,0]
```

```python
#2) 1에서 가져온 이름 정보 값이 '민철'인지 아닌지 판멸
name_score[ :, 0] == '민철'
```

```python
#3) '민철'의 정보를 가져오기 => 불리언 인덱싱 
name_score[name_score[ :, 0] == '민철']
```

```python
#점수가 80점 이하인 정보
#1) 점수 정보만 가져오기 
name_score [ :,1]

#문자를 숫자로 먼저 바꿔줘야 함. 

#3) 점수가 80점 이하 조건 => 불리언 인덱싱 
```

```python
#문자를 숫자로 먼저 바꿔줘야 함. 
name_score [ :,1].astype('int') <= 80
#2) 1에서 가져온 점수 정보 값이 80 이하인지 판별
```

```python
#3) 점수가 80점 이하 조건 => 불리언 인덱싱 
name_score[name_score [ :,1].astype('int') <= 80]
```
