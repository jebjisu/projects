- 교육수준에 따른 평균 연봉 확인 

```python
#교육수준('EDU_LEVEL')의 종류 확인 
hr_info['EDU_LEVEL'].unique()
```

```python
#'박사 학위'인 직원의 연봉(SALARY) 값의 평균
pdm = hr_info[hr_info['EDU_LEVEL'] == '박사 학위']
pdm
```

```python
#연봉컬럼만 가져와서 평균 계산
pdm['SALARY'].mean()
```

```python
# 교육수준 종류에 따른 평균 연봉 모두 확인 
for el in hr_info['EDU_LEVEL'].unique():
    pdm = hr_info[hr_info['EDU_LEVEL'] == el]
    sal_mean = pdm['SALARY'].mean()
    print(f"{el}:{round(sal_mean,2)}")

 
```

- 데이터 그룹화를 이용한 통계분석
   - groupby: 카테고리별 그룹을 만들어 연산 수행

```python
# 교육수준 종류에 따른 평균
hr_info[ ['EDU_LEVEL', 'SALARY'] ].groupby(by='EDU_LEVEL').mean() #데이터를 가지고 올 데이터 
#출력하고자 하는 카테고리가 기준 -> by
```

```python
#교육수준과 성별에 따른 평균 연봉 
hr_info[ ['EDU_LEVEL', 'GENDER','SALARY'] ].groupby(by=['EDU_LEVEL','GENDER']).mean()
```

```python
#교육수준과 성별 별로 가장 많은 연봉을 받는 사람이 궁금
hr_info[ ['EDU_LEVEL', 'GENDER','SALARY'] ].groupby(by=['EDU_LEVEL','GENDER']).agg(['max','min','median'])
```
