```python
#hr_info로 groupby 활용하기
#조직별 인원수, 평균 나이, 평균근속연수, 평균연봉 확인 
#ENAME, AGE, SERVICE_YEAR, SALARY 
#기준값 : ORC_CD
org_agg = hr_info.groupby(by = 'ORG_CD').agg({'ENAME':'count', 
                                              'AGE':'mean', 
                                              'SERVICE_YEAR':'mean', 
                                              'SALARY':'mean'})
```

```python
#컬럼명 수정
org_agg.columns = ['HEAD_CNT', 'AGE_MEAN', 'SERVICE_YEAR_MEAN', 'SALARY_MEAN']
```

```python
org_agg.head(2)
```

```python
#두 df모두 인덱스에 조직 정보가 있음
#인덱스 기준으로 병합  -> axis=1
org_res = pd.concat([org_info, org_agg ],axis = 1)
```

```python
org_res.to_csv('./data/org_res.csv', index = False, encoding = 'EUC-KR')
```

### DataFrame 병합2
- pd.merge( df1, df2, how - "JOIN속성", on="key값 (column)")
- 두 DF에서 공통된 키 값을 기준으로 병합 (데이터를 기준으로 병합)
- JOIN 속성
   - inner : 교집합 (기본)
   - outer : 합집합 병합
   - left : 왼쪽 데이터 기준 
   - right : 오른쪽 df 기준
  

#### concat, merge
- concat
   - 병합기준 : 인덱스명, 컬럼명
   - 단순연결
- merge
   - 병합기준 : 데이터값(공통된 키값)
   - 데이터셋의 관계를 바탕으로 결합
   - 검색기준 : JOIN

```python
hr_info_pay = pd.read_csv('./data/hr-pay-info.csv',  encoding = 'EUC-KR')
hr_info_pay
```

```python
#score 파일 읽어오기 
score = pd.read_csv('./data/score.csv', encoding='EUC-KR', index_col='과목' )
score
```

```python
# 1. 과목별 합계를 구해서 DF에 추가('합계' 라는 컬럼 추가)

#과목 합계 sum axis = 1
# 과목별 연산
#데이터 프레임에 컬럼 추가를 해야 함. 
#score['합계']= score.sum(axis = 1)
#score['합계']= score[['1반','2반','3반','4반']].sum(axis=1)
#score['합계']=score.loc[ :, '1반':'4반'].sum(axis=1)
score['합계']=score.iloc[ :, 0:4].sum(axis=1)
score
```

```python
#2. 과목별 평균을 계산해서 '평균이라는 컬럼을 만들기 
```

```python
score['평균']=score.iloc[:, 0:4].mean(axis=1)
score
```

```python
score=score.drop('평균',axis=1)
```

```python
score['평균']=score.loc[ : , '1반':'4반'].mean(axis=1)
score
#평균은 인덱싱을 활용함...
```

```python
#인덱싱으로 반평균 데이터만 계산 (합계, 평균x)
score.loc['반평균']=score.loc['파이썬':'Web', '1반':'4반'].mean(axis=0)

score
```

```python
#결측치를'-' 로 채우기 
score.fillna('-', inplace = True)
score
```

```python
#4. 과목별 가장 높은 점수와 가장 낮은 점수의 차이 구하기

#가장 높은 점수 max
max_socres = score.loc['파이썬':'Web' , '1반':'4반'].max(axis=1)
max_socres
```

```python
#가장 낮은 점수 min
min_socres = score.loc['파이썬':'Web' , '1반':'4반'].min(axis=1)

min_socres
```

```python
max_socres - min_socres
```

```python
def max_min(x):
    return x.max() - x.min()
```

```python
score.loc[ :'Web','1반':'4반'].apply(max_min, axis=1)
```
