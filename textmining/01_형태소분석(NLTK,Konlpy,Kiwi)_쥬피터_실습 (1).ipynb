##### ♦️1. 실습예제
  - 명사만 추출하는 define 함수를 만들어보자!

```python
def nltk_noun(string):
    #품사 분리해주기 
    split_text = nltk.word_tokenize(string)
    tagged_text = nltk.pos_tag(split_text)

    # #for문
    # result = [] # 명사만 들어갈거임 
    # for word, tag in tagged_text:
    #     #첫번째 태그가 명사면 result에 담기 
    #     if tag in {'NN', 'NNS','NNP','NNPS'}:
    #         result.append(word)

    #컨프레이션 사용하기 
    result = [word for word, tag in tagged_text if tag in {'NN', 'NNS','NNP','NNPS'}] #어떤 요소를 변수에 채울지 (word)

    return result
```

```python
nltk_noun('Hello My name is Jisu')
```

##### ♦️2. 실습예제
  - 영어 불용어 적용 명사 define 만들기

```python


def extract_noun_english(string):
    # string = string.lower()
    # split_text = nltk.word_tokenize(string)
    # tagged_text = nltk.pos_tag(split_text)
    
    # result = [word for word, tag in tagged_text if word not in stopwords if tag in {'NN', 'NNS','NNP','NNPS'}]

    result = []
    for word, tag in tagged_text:
        if word not in stopwords: #사전에 안들은 것 중에 즉, 불용어가 아닌 것 중에 
            if tag in {'NN', 'NNS','NNP','NNPS'}:
                result.append(word)

                
    return result
    
```

```python
from konlpy.tag import Okt
okt=Okt()
```

```python
text = '오늘 나는 화육계에 갈 것이다.'
okt.morphs(text)
```

```python
#정규화
text = '맛있는 닭발을 먹을겅밈ㅋㅋㅋㅋㅋㅋㅋㅋ'
okt.pos(text,norm=True)
```

```python
#어간추출
text = '퇴근 하자마자 달려갈 것이다 !!!!.'
okt.morphs(text,stem = True)
```

```python
text = '배고프다. 매우 배고프다. 퇴근 하고싶다.'
okt.morphs(text, norm= True, stem = True)
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
kiwi.tokenize("아버지가 방에 들어가신다")
```

```python
#원하는 형태소 뽑기
result = [i[0] for i in kiwi.tokenize("아버지가 방에 들어가신다") if i[1] in {'NNG'}]
result
```
