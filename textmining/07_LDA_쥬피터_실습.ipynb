```python
#ë¶ˆìš©ì–´ ê°€ì ¸ì˜¤ê¸° 
stopwords = nltk.corpus.stopwords.words('english')
```

```python
import re
```

```python
#ì˜ì–´ ì•ŒíŒŒë²³ë§Œ ë‚¨ê¸°ê³  ë‹¤ ì§€ìš¸ê±°ì„ 
# ëª…ì‚¬ ë™ì‚¬ë§Œ ë½‘ê³ 
#ë¶ˆìš©ì–´ ì œê±° 
def pos_tagging(string):
    #íŠ¹ìˆ˜ë¬¸ì ì œê±°í•˜ê¸°(ì˜ì–´ ì•ŒíŒŒë²³ë§Œ ë‚¨ê¸°ê¸°)
    eng = re.compile('[^a-zA-Z]')
    string = re.sub(eng, ' ', string) #ì»´íŒŒì¼ íŒ¨í„´ê°’/ ë„ì–´ì“°ê¸°ë¡œ ë°”ê¿”ì¤Œ, (ë‰´ìŠ¤ê¸°ì‚¬ê°€ ì•„ë‹ˆë¼ì„œ ì‚­ì œí•˜ëŠ” ê²ƒë³´ë‹¤ ëŠë‚Œí‘œë¥¼ í•˜ëŠ”ê²Œ ì¢‹ìŒ), ê³µë°± í•˜ë‚˜ë¡œ ì¹˜í™˜í•˜ê¸° 

    eng2 = re.compile('\s+') #ì—¬ëŸ¬ê°œ ê³µë°±ì„
    string = re.sub(eng2, ' ', string) # í•˜ë‚˜ë¡œ ì¤„ì—¬ì¤Œ 

    #ë¶ˆìš©ì–´ ì²˜ë¦¬ 
    #1. ì†Œë¬¸ì ë³€í™˜í•˜ê¸° 
    strimg = string.lower()
    string = nltk.word_tokenize(string)
    string = nltk.pos_tag(string)

    #í’ˆì‚¬ë½‘ê¸° 
    result = [word for word, tag in string if word not in stopwords if tag in {'NN','NNS','NNP','NNPS', 'VB', 'VBD','VBG','VBN','VBP','VBZ'}] #string ë‹¤ìŒì—ëŠ” ì¡°ê±´ë¬¸ if
    return result 
```

```python
import gensim 
from gensim import corpora, models
```

```python
dictionary = corpora.Dictionary(tagged_pos_document)
```

```python
#ID ë‘ ë§¤ì¹­ 

corpus = [dictionary.doc2bow(doc) for doc in tagged_pos_document ]
corpus
```

## 3. LDA í† í”½ ìˆ˜ ì„¤ì •í•˜ê¸°

```python
top_n = 3
#ëª¨ë¸ ìƒì„±
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=top_n, id2word=dictionary)
```

```python
ldamodel.print_topics(num_words = 10)
```

## 4. LDA í† í”½ ìˆ˜ ì°¾ê¸°
- Coherence & perplexity
  * Coherence : 0~1 (í´ìˆ˜ë¡ êµ³)
  * Perplexity : ìŒìˆ˜ì¼ìˆ˜ë¡ (ì‘ì„ìˆ˜ë¡ êµ³) 

```python
#coherence
from gensim.models import CoherenceModel
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=top_n, id2word=dictionary)
#ëª¨ë¸ ì„ ì–¸í•´ì£¼ê¸° 
coherencemodel = CoherenceModel(model = ldamodel,
                                texts = tagged_pos_document,
                               dictionary = dictionary)
```

```python
coherence_score = coherencemodel.get_coherence()
```

```python
#perplexity
ldamodel.log_perplexity(corpus)
```

```python
#ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
import matplotlib.pyplot as plt
```

```python
from tqdm import tqdm

for i in tqdm(range(2, 10)):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=top_n, id2word=dictionary)

    coherencemodel = CoherenceModel(model = ldamodel,
                                texts = tagged_pos_document,
                               dictionary = dictionary)

    c_score.append(coherencemodel.get_coherence())

    #perplexity
    p_score.append(ldamodel.log_perplexity(corpus))
```

### 4.1 Coherence

```python
x = range(2, 10)
plt.plot(x, c_score)
plt.show()
```

### 4.2 Perplexity

```python
x = range(2,10)
plt.plot(x, p_score)
plt.show()
```
