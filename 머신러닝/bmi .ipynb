```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import warnings
warnings.filterwarnings(action = 'ignore')

df = pd.read_csv("C:/Users/152/Downloads/bmi_500.csv")
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
#훈련변수 #평가세트 #훈련변수의 종속변수 #평가세트의 종속변수
x_train, x_test, y_train, y_text = train_test_split(df[["Weight","Height"]],df["Label"], random_state = 42)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression #숫자만 받을 수 있음 (인코딩 필수)
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

#모델은
model = DecisionTreeClassifier(max_depth = 10, max_leaf_nodes=20)
# model = RandomForestClassifier() #랜덤포레스트 모델 
#모델 학습
model.fit(x_train, y_train)
```

```python
#예측하기 
model.predict(x_train)
```

```python
#얼마나 잘했는지 점수를 매기기 
model.score(x_train, y_train)
```

```python
model.score(x_test, y_text)
```
